# **Data Acquisition - Web Scraping**

```python
import requests
import json
import pandas as pd
```

```python
payload={
   "requests":[
      {
         "indexName":"job_postings",
         "params":"query=&hitsPerPage=1000&maxValuesPerFacet=1000&page=0&facets=%5B%22*%22%2C%22city.work_country_name%22%2C%22position.name%22%2C%22industries.vertical_name%22%2C%22experience%22%2C%22job_type.name%22%2C%22is_salary_visible%22%2C%22has_equity%22%2C%22currency.currency_code%22%2C%22salary_min%22%2C%22taxonomies.slug%22%5D&tagFilters=&facetFilters=%5B%5B%22city.work_country_name%3AIndonesia%22%5D%5D"
      },
      {
         "indexName":"job_postings",
         "params":"query=&hitsPerPage=1&maxValuesPerFacet=1000&page=0&attributesToRetrieve=%5B%5D&attributesToHighlight=%5B%5D&attributesToSnippet=%5B%5D&tagFilters=&analytics=false&clickAnalytics=false&facets=city.work_country_name"
      }
   ]
}

headers={'content-type': 'application/x-www-form-urlencoded'}

api_url = "https://219wx3mpv4-dsn.algolia.net/1/indexes/*/queries?x-algolia-agent=Algolia%20for%20vanilla%20JavaScript%203.30.0%3BJS%20Helper%202.26.1&x-algolia-application-id=219WX3MPV4&x-algolia-api-key=b528008a75dc1c4402bfe0d8db8b3f8e"

jsonData=requests.post(api_url,data=json.dumps(payload),headers=headers).json()
#print(jsonData)

job_info_df = pd.DataFrame()
temp_dict = {}

for item in jsonData['results'][0]['hits']:
    temp_dict['Job Title'] = item['_highlightResult']['title']['value']
    temp_dict['Job Desc'] = item['description']
    temp_dict['Job Type'] = item['job_type']['name']
    temp_dict['Experience'] = item['experience'] 
    temp_dict['Is Remote'] = item['is_remote']
    temp_dict['Position'] = item['position']['name']
    temp_dict['Vacancy Count'] = item['vacancy_count']
    temp_dict['Company'] = item['_highlightResult']['company']['name']['value']
    temp_dict['Industries'] = item['industries'][0]['name']
    temp_dict['City']= item['_highlightResult']['city']['name']['value']
    temp_dict['Skill'] = item['_highlightResult']['job_skills'][0]['name']['value']
    temp_dict['Salary Max'] = item['salary_max']
    temp_dict['Salary Min'] = item['salary_min']
    temp_dict['Published At'] = item['published_at']
    temp_dict['Expires At'] = item['expires_at']

    job_info_df = job_info_df.append(temp_dict, ignore_index=True)
    job_info_df.to_csv('TechinAsia Dataset.csv', index=False)
    print(job_info_df)
```
